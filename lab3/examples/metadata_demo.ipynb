## Generate some fake metadata ##

```python
import numpy as np
fake_labels2 = pd.DataFrame([sigs.index.astype(object), ['tuned' if int(10*np.random.rand()) % 2 else 'untuned' for i in sigs.index]], index=['roi_label', 'spatially_tuned']).T
fake_labels2.to_csv(os.path.join(sima_path, 'fake_labels2.csv'))
fake_labels2
```
